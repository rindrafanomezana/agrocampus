```julia
using Random, Plots

n = 2000
seuil = 0.25
rng = MersenneTwister(1234)
X1 = rand( rng, n)
X2 = rand( rng, n)
U  = rand( rng, n)
Y  = zeros(Int32,n)
Y[(X1 .<= 0.25) .& (U  .<= seuil)] .= 1
Y[(X1 .>  0.25) .& (X2 .>= 0.75) .& (U .<= seuil)] .= 1
Y[(X1 .>  0.25) .& (X2 .<  0.75) .& (U .>  seuil)] .= 1

scatter(X1,X2, marker_z = Y)
```
